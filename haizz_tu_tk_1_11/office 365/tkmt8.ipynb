Câu 1. (6 điểm) Cho ma trận $A=\begin{bmatrix}1 & 0\\ 0 & -4 \end{bmatrix}$ và vector $b=(2;2)$. Xét bài toán tối ưu sau 
$$\min_{x\in\mathbb{R}^2}f(x)=\dfrac{1}{2}\|Ax-b\|_2^2.$$

a) (2 điểm) Xác định điểm tối ưu $x^*$ và giá trị tối ưu $p^*$ của bài toán trên.

b) (2 điểm) Sử dụng thuật toán Gradient Descent, với giá trị $x$ ban đầu là $x^{(0)}=(-1;2)$, sử dụng learning rate lần lượt là $\dfrac{1}{7}$ và $\dfrac{1}{10}$ và thực hiện tối đa $100$ vòng lặp. In ra giá trị của $x^{(k)}$, $f(x^{(k)})$ tương ứng sau mỗi vòng lặp $k$ và vẽ đồ thị biểu thị cho sai số $|f(x^{(k)})-p^*|$ trong cả hai trường hợp của learning rate. Từ đó hãy đưa ra kết luận về sự hội tụ của thuật toán Gradient Descent trong từng trường hợp của learning rate.

c) (1 điểm) Tìm điều kiện cụ thể cho learning rate để thuật toán Gradient Descent của bài toán trên hội tụ.

d) (1 điểm) Trả lời câu a) với ma trận $A=\begin{bmatrix}1 & -2\\ 2 & -4 \end{bmatrix}$ và vector $b=(2;2)$.


```python
import autograd.numpy as np
from autograd import grad
import matplotlib.pyplot as plt
```

```python
def create_least_squares_problem(A, b):
    def f(x):
        r = A @ x - b
        return 0.5 * np.dot(r, r)
    return f
```

```python
def gradient_descent(f, grad_f, x0, lr, max_iter):
    x = x0.copy()
    xs, fs = [x.copy()], [f(x)]

    print(f"\n=== Gradient Descent (lr = {lr}) ===")
    for k in range(max_iter):
        x = x - lr * grad_f(x)
        fx = f(x)
        xs.append(x.copy())
        fs.append(fx)

        print(f"Iter {k+1:3d}: x = {x}, f(x) = {fx:.6f}")

    return np.array(xs), np.array(fs)
```

```python
def gd_convergence_condition(A):
    """
    Với f(x)=1/2||Ax-b||^2
    Hessian = A^T A
    GD hội tụ nếu 0 < lr < 2 / lambda_max(A^T A)
    """
    H = A.T @ A
    eigvals = np.linalg.eigvals(H)
    L = np.max(eigvals).real
    return 2 / L
```

```python
if __name__ == "__main__":

    # ==================================================
    # INPUT – BẠN CHỈ CẦN ĐỔI PHẦN NÀY
    # ==================================================
    A = np.array([[1., 0.],
                  [0., -4.]])
    b = np.array([2., 2.])
    x0 = np.array([-1., 2.])

    lr_list = [1/7, 1/10]
    max_iter = 100

    # ==================================================
    # (a) Tìm nghiệm tối ưu
    # ==================================================
    f = create_least_squares_problem(A, b)
    grad_f = grad(f)

    # Nghiệm tối ưu (least squares)
    x_star = np.linalg.pinv(A) @ b
    p_star = f(x_star)

    print("===== (a) NGHIỆM TỐI ƯU =====")
    print("x* =", x_star)
    print("p* =", p_star)

    # ==================================================
    # (b) Gradient Descent với 2 learning rate
    # ==================================================
    histories = {}

    for lr in lr_list:
        xs, fs = gradient_descent(f, grad_f, x0, lr, max_iter)
        histories[lr] = (xs, fs)

    # ==================================================
    # Vẽ đồ thị sai số
    # ==================================================
    plt.figure(figsize=(8, 5))

    for lr in lr_list:
        _, fs = histories[lr]
        error = np.abs(fs - p_star)
        plt.semilogy(error, label=f"lr = {lr}")

    plt.xlabel("Iteration k")
    plt.ylabel(r"$|f(x^{(k)}) - p^*|$")
    plt.title("Sai số Gradient Descent")
    plt.grid(True)
    plt.legend()
    plt.show()

    # ==================================================
    # (c) Điều kiện hội tụ
    # ==================================================
    lr_max = gd_convergence_condition(A)

    print("\n===== (c) ĐIỀU KIỆN HỘI TỤ =====")
    print("Gradient Descent hội tụ nếu:")
    print("0 < learning rate < ", lr_max)

    # ==================================================
    # (d) Đổi A, b
    # ==================================================
    A2 = np.array([[1., -2.],
                   [2., -4.]])
    b2 = np.array([2., 2.])

    f2 = create_least_squares_problem(A2, b2)
    x_star_2 = np.linalg.pinv(A2) @ b2
    p_star_2 = f2(x_star_2)

    print("\n===== (d) TRƯỜNG HỢP MA TRẬN MỚI =====")
    print("x* =", x_star_2)
    print("p* =", p_star_2)
```

Câu 2. (4 điểm) Cho ma trận $A=\begin{bmatrix}1 & 0 & 0\\ 0 & 2 & 0 \\ 0 & 0 & 3 \end{bmatrix}$ và vector $b=(1/2;1; 3/2)$. Xét bài toán tối ưu sau $\begin{align}
\min_{x=(x_1,x_2,x_3)\in\mathbb{R}^3}\;\;\; &f_0(x)=\sum_{i=1}^3x_i\log x_i\\
\text{subject to }\;\;\; & Ax\geq b, \quad\quad\quad\quad \quad (1) \\
& \mathbf{1}^Tx=\dfrac{3}{2}.
\end{align}$

a) (1 điểm) Xác định hàm Lagrange $L(x,\lambda,v)$ cho bài toán $(1)$.

b) (2 điểm) Gọi $x^*$ và $(\lambda^*, v^*)$ lần lượt là điểm tối ưu của bài toán $(1)$ và bài toán đối ngẫu của nó. Viết điều kiện Karush-Kuhn-Tucker (KKT) cho $x^*$ và $(\lambda^*, v^*)$.

c) (1 điểm) Tìm một giá trị của $x^*$ và $(\lambda^*, v^*)$ thoả điều kiện KKT vừa tìm được ở câu b), từ đó đưa ra giá trị tối ưu cho bài toán $(1)$.

Lưu ý: sinh viên hoàn thành Câu 2 theo mẫu dưới đây.

a) Hàm Lagrange $L(x,\lambda,v)$

 
a) Hàm Lagrange $L(x,\lambda,v) = f_0(x) + \lambda^T(Ax - b) + v(\mathbf{1}^Tx - \frac{3}{2})$.



```python
import numpy as np
import cvxpy as cp
import sympy as sp

# =========================================================
# (A) Sinh biểu thức Lagrange L(x, λ, v) dạng ký hiệu (sympy)
# =========================================================
def build_lagrangian_symbolic(A, b, s):
    """
    A: (m,n), b: (m,), s: scalar for 1^T x = s
    Return: sympy expression L(x, lambda, v)
    """
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)
    m, n = A.shape

    x = sp.Matrix(sp.symbols(f"x1:{n+1}", positive=True, real=True))  # x_i > 0 for log
    lam = sp.Matrix(sp.symbols(f"lambda1:{m+1}", nonnegative=True, real=True))  # λ >= 0
    v = sp.Symbol("v", real=True)

    A_sp = sp.Matrix(A)
    b_sp = sp.Matrix(b)

    f0 = sum(x[i] * sp.log(x[i]) for i in range(n))
    L = f0 + (lam.T * (A_sp * x - b_sp))[0] + v * ((sp.Matrix([1]*n).T * x)[0] - s)
    return x, lam, v, sp.simplify(L)

# =========================================================
# (B) In điều kiện KKT (dạng công thức + kiểm tra số)
# =========================================================
def print_kkt_conditions_symbolic(A, b, s):
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)
    m, n = A.shape

    # Khai báo biến symbolic
    x = sp.Matrix(sp.symbols(f"x1:{n+1}", positive=True, real=True))
    lam = sp.Matrix(sp.symbols(f"lambda1:{m+1}", nonnegative=True, real=True))
    v = sp.Symbol("v", real=True)

    A_sp = sp.Matrix(A)
    b_sp = sp.Matrix(b)
    ones = sp.Matrix([1]*n)

    print("\n===== (b) ĐIỀU KIỆN KKT (CỤ THỂ) =====")

    # 1. Primal feasibility
    print("\n1) Primal feasibility:")
    for i in range(m):
        expr = (A_sp.row(i) * x)[0]   # scalar
        print(f"   {sp.pretty(expr)}  ≥  {b[i]}")
    print(f"   {sp.pretty((ones.T * x)[0])} = {s}")
    for i in range(n):
        print(f"   x{i+1} > 0")

    # 2. Dual feasibility
    print("\n2) Dual feasibility:")
    for i in range(m):
        print(f"   λ{i+1} ≥ 0")

    # 3. Complementary slackness  ✅ FIX Ở ĐÂY
    print("\n3) Complementary slackness:")
    for i in range(m):
        slack = (A_sp.row(i) * x)[0] - b_sp[i, 0]   # scalar − scalar
        print(f"   λ{i+1} · ({sp.pretty(slack)}) = 0")

    # 4. Stationarity
    print("\n4) Stationarity (∂L/∂x = 0):")
    grad_f = sp.Matrix([sp.log(x[i]) + 1 for i in range(n)])
    station = grad_f + A_sp.T * lam + v * ones

    for i in range(n):
        print(f"   ∂L/∂x{i+1} = {sp.pretty(station[i])} = 0")

# =========================================================
# (C) Giải primal bằng CVXPY và lấy (x*, λ*, v*)
# =========================================================
def solve_entropy_problem(A, b, s, solver=None, verbose=False):
    """
    Minimize sum_i x_i log x_i
    s.t. Ax >= b, 1^T x = s, x >= 0

    Trả về x_star, lambda_star (dual for Ax>=b), v_star (dual for equality)
    """
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)
    m, n = A.shape

    x = cp.Variable(n, nonneg=True)

    # Objective: sum x log x is convex on x>0.
    # Use KL divergence trick: kl_div(x,1) = x*log(x) - x + 1
    # => sum x log x = sum kl_div(x,1) + sum x - n (constant shift since sum x = s)
    obj = cp.Minimize(cp.sum(cp.kl_div(x, np.ones(n))))

    cons_ineq = (A @ x >= b)
    cons_eq = (cp.sum(x) == s)
    constraints = [cons_ineq, cons_eq]

    prob = cp.Problem(obj, constraints)

    # Solve
    if solver is None:
        prob.solve(verbose=verbose)
    else:
        prob.solve(solver=solver, verbose=verbose)

    x_star = x.value
    # Dual variables:
    lambda_star = cons_ineq.dual_value   # shape (m,)
    v_star = cons_eq.dual_value          # scalar

    # Tính đúng p* = sum x log x (không dùng obj đã shift)
    eps = 1e-12
    p_star = float(np.sum(x_star * np.log(np.maximum(x_star, eps))))

    return x_star, lambda_star, float(v_star), p_star, prob.status

# =========================================================
# (D) Kiểm tra KKT bằng số cho nghiệm tìm được
# =========================================================
def check_kkt_numeric(A, b, s, x_star, lambda_star, v_star, tol=1e-7):
    A = np.asarray(A, dtype=float)
    b = np.asarray(b, dtype=float).reshape(-1)
    x_star = np.asarray(x_star, dtype=float).reshape(-1)
    lambda_star = np.asarray(lambda_star, dtype=float).reshape(-1)

    # Primal residuals
    r_ineq = A @ x_star - b
    r_eq = np.sum(x_star) - s

    # Dual feasibility
    dual_ok = np.all(lambda_star >= -tol)

    # Complementary slackness
    comp = lambda_star * r_ineq
    comp_ok = np.all(np.abs(comp) <= 1e-5)

    # Stationarity: log(x)+1 + A^T λ + v*1 = 0
    eps = 1e-12
    station = (np.log(np.maximum(x_star, eps)) + 1.0) + (A.T @ lambda_star) + v_star * np.ones_like(x_star)
    station_ok = np.all(np.abs(station) <= 1e-5)

    primal_ok = (np.all(r_ineq >= -tol) and abs(r_eq) <= 1e-5 and np.all(x_star >= -tol))

    return {
        "primal_ok": primal_ok,
        "dual_ok": dual_ok,
        "comp_ok": comp_ok,
        "station_ok": station_ok,
        "ineq_residual": r_ineq,
        "eq_residual": r_eq,
        "comp_residual": comp,
        "station_residual": station
    }

# =========================================================
# MAIN: chạy đúng bài bạn đưa
# =========================================================
if __name__ == "__main__":
    # ----- INPUT (chỉ cần đổi chỗ này cho bài khác) -----
    A = np.array([
        [1., 0., 0.],
        [0., 2., 0.],
        [0., 0., 3.]
    ])
    b = np.array([0.5, 1.0, 1.5])
    s = 1.5  # 1^T x = 3/2

    # (a) Sinh & in Lagrangian theo mẫu
    x_sym, lam_sym, v_sym, L_sym = build_lagrangian_symbolic(A, b, s)
    print("===== (a) HÀM LAGRANGE =====")
    print("L(x, λ, v) =")
    print(L_sym)

    # (b) In điều kiện KKT
    print_kkt_conditions_symbolic(A, b, s)


    # (c) Tìm một nghiệm thỏa KKT bằng cách giải primal + lấy dual
    x_star, lambda_star, v_star, p_star, status = solve_entropy_problem(A, b, s)
    print("\n===== (c) NGHIỆM THỎA KKT (từ nghiệm tối ưu primal/dual) =====")
    print("Solver status:", status)
    print("x* =", x_star)
    print("lambda* =", lambda_star)
    print("v* =", v_star)
    print("p* = sum_i x_i log x_i =", p_star)

    # Check KKT numeric
    kkt = check_kkt_numeric(A, b, s, x_star, lambda_star, v_star)
    print("\n===== KIỂM TRA KKT (NUMERIC) =====")
    print("Primal feasible:", kkt["primal_ok"])
    print("Dual feasible  :", kkt["dual_ok"])
    print("Complementary  :", kkt["comp_ok"])
    print("Stationarity   :", kkt["station_ok"])
    print("Ineq residual Ax-b:", kkt["ineq_residual"])
    print("Eq residual 1^T x - s:", kkt["eq_residual"])
```

```python
# ============================================================
# KIẾN TRÚC CHUẨN KKT (TỔNG QUÁT n-CHIỀU)
#   - Chuẩn hóa ràng buộc về dạng:
#       g(x) <= 0
#       h(x) = 0
#   - Sinh Lagrangian L = f + λ^T g + v h
#   - Sinh KKT (primal/dual feasibility, complementary slackness, stationarity)
#   - Giải primal/dual bằng CVXPY và kiểm tra KKT bằng số
# ============================================================

import numpy as np
import cvxpy as cp
import sympy as sp
from dataclasses import dataclass

# -----------------------------
# 1) Đặc tả bài toán (linh hoạt)
# -----------------------------
@dataclass
class KKTProblemSpec:
    A: np.ndarray          # (m,n)
    b: np.ndarray          # (m,)
    c: np.ndarray          # (n,)  (vector trong đẳng thức c^T x = s)
    s: float               # scalar
    ineq_type: str = ">="  # ">=" hoặc "<=" ; nghĩa là Ax >= b hoặc Ax <= b

    def __post_init__(self):
        self.A = np.asarray(self.A, dtype=float)
        self.b = np.asarray(self.b, dtype=float).reshape(-1)
        self.c = np.asarray(self.c, dtype=float).reshape(-1)
        m, n = self.A.shape
        assert self.b.shape == (m,), "b phải có shape (m,)"
        assert self.c.shape == (n,), "c phải có shape (n,)"
        assert self.ineq_type in (">=", "<="), "ineq_type chỉ nhận '>=' hoặc '<='"

    # ----------------------------------------
    # 2) Chuẩn hóa ràng buộc về g(x) <= 0, h(x)=0
    # ----------------------------------------
    def g_linear(self, x):
        """g(x) <= 0 (numpy)"""
        if self.ineq_type == ">=":
            # Ax >= b  <=>  b - Ax <= 0
            return self.b - self.A @ x
        else:
            # Ax <= b  <=>  Ax - b <= 0
            return self.A @ x - self.b

    def h_linear(self, x):
        """h(x)=0 (numpy)"""
        return float(self.c @ x - self.s)

    def Jg(self):
        """Jacobian của g theo x: g(x)=M x + q => Jg = M"""
        if self.ineq_type == ">=":
            # g = b - A x => Jg = -A
            return -self.A
        else:
            # g = A x - b => Jg = A
            return self.A

    def Jh(self):
        """Jacobian của h theo x: h = c^T x - s => Jh = c^T"""
        return self.c.reshape(1, -1)

# -----------------------------------------
# 3) Objective entropy: f(x)=sum x_i log x_i
# -----------------------------------------
def f_entropy_np(x, eps=1e-12):
    x = np.asarray(x, dtype=float)
    return float(np.sum(x * np.log(np.maximum(x, eps))))

def build_symbolic_objects(spec: KKTProblemSpec):
    """Sinh x, λ, v và biểu thức L(x,λ,v) theo SymPy (tổng quát)."""
    m, n = spec.A.shape

    x = sp.Matrix(sp.symbols(f"x1:{n+1}", positive=True, real=True))  # x_i > 0
    lam = sp.Matrix(sp.symbols(f"lambda1:{m+1}", nonnegative=True, real=True))  # λ >= 0
    v = sp.Symbol("v", real=True)

    A_sp = sp.Matrix(spec.A)
    b_sp = sp.Matrix(spec.b)
    c_sp = sp.Matrix(spec.c)

    # f(x) = Σ x log x
    f0 = sum(x[i] * sp.log(x[i]) for i in range(n))

    # g(x) <= 0
    if spec.ineq_type == ">=":
        # g = b - A x
        g = b_sp - A_sp * x
    else:
        # g = A x - b
        g = A_sp * x - b_sp

    # h(x)=0: c^T x - s
    h = (c_sp.T * x)[0] - spec.s

    # L = f + λ^T g + v h
    L = f0 + (lam.T * g)[0] + v * h
    return x, lam, v, sp.simplify(L), g, h

# -----------------------------------------
# 4) In bài làm: Lagrangian + KKT "cụ thể"
# -----------------------------------------
def print_lagrangian_and_kkt(spec: KKTProblemSpec):
    x, lam, v, L, g_sym, h_sym = build_symbolic_objects(spec)
    m, n = spec.A.shape

    print("===== (a) HÀM LAGRANGE =====")
    print("L(x, λ, v) =")
    print(L)

    print("\n===== (b) ĐIỀU KIỆN KKT (CỤ THỂ – TỰ ĐỘNG THEO INPUT) =====")

    # 1) Primal feasibility
    print("\n1) Primal feasibility:")
    # In ra ràng buộc g(x) <= 0 tương ứng với Ax >= b / Ax <= b
    for i in range(m):
        if spec.ineq_type == ">=":
            # Ax >= b
            lhs = (sp.Matrix(spec.A).row(i) * sp.Matrix(x))[0]
            print(f"   {sp.pretty(lhs)}  ≥  {spec.b[i]}")
        else:
            lhs = (sp.Matrix(spec.A).row(i) * sp.Matrix(x))[0]
            print(f"   {sp.pretty(lhs)}  ≤  {spec.b[i]}")
    # Equality
    c_sp = sp.Matrix(spec.c)
    print(f"   {sp.pretty((c_sp.T * x)[0])}  =  {spec.s}")
    # Domain
    for i in range(n):
        print(f"   x{i+1} > 0")

    # 2) Dual feasibility
    print("\n2) Dual feasibility:")
    for i in range(m):
        print(f"   λ{i+1} ≥ 0")

    # 3) Complementary slackness: λ_i * g_i(x)=0
    print("\n3) Complementary slackness:")
    for i in range(m):
        print(f"   λ{i+1} · ({sp.pretty(g_sym[i])}) = 0")

    # 4) Stationarity: ∇f + Jg^T λ + Jh^T v = 0
    print("\n4) Stationarity (∂L/∂x = 0):")
    grad_f = sp.Matrix([sp.log(x[i]) + 1 for i in range(n)])
    Jg_sp = sp.Matrix(spec.Jg())          # (m,n)
    Jh_sp = sp.Matrix(spec.Jh())          # (1,n)
    station = grad_f + Jg_sp.T * lam + Jh_sp.T * v
    for i in range(n):
        print(f"   ∂L/∂x{i+1} = {sp.pretty(sp.simplify(station[i]))} = 0")

# -----------------------------------------
# 5) Giải primal/dual bằng CVXPY (tổng quát)
# -----------------------------------------
def solve_primal_dual(spec: KKTProblemSpec, solver=None, verbose=False):
    """
    Giải:
        min Σ x_i log x_i
        s.t. Ax (>= or <=) b
             c^T x = s
             x >= 0
    Lấy dual λ* cho bất đẳng thức (tương ứng g(x)<=0 chuẩn hóa) và v* cho đẳng thức.
    """
    A, b, c, s = spec.A, spec.b, spec.c, spec.s
    m, n = A.shape

    x = cp.Variable(n, nonneg=True)

    # Dùng kl_div: kl_div(x,1) = x log x - x + 1 (convex)
    # Vì c^T x = s cố định, nên tối ưu tương đương
    obj = cp.Minimize(cp.sum(cp.kl_div(x, np.ones(n))))

    if spec.ineq_type == ">=":
        cons_ineq = (A @ x >= b)
    else:
        cons_ineq = (A @ x <= b)

    cons_eq = (c @ x == s)

    prob = cp.Problem(obj, [cons_ineq, cons_eq])

    if solver is None:
        prob.solve(verbose=verbose)
    else:
        prob.solve(solver=solver, verbose=verbose)

    x_star = np.asarray(x.value, dtype=float).reshape(-1)

    # CVXPY dual_value trả về dual đúng theo canonical g(x)<=0:
    # - Nếu viết A x >= b => canonical là b - A x <= 0 => dual >=0 chính là λ cho g=b-Ax
    # - Nếu viết A x <= b => canonical là A x - b <= 0 => dual >=0 chính là λ cho g=Ax-b
    lambda_star = np.asarray(cons_ineq.dual_value, dtype=float).reshape(-1)
    v_star = float(cons_eq.dual_value)

    p_star = f_entropy_np(x_star)

    return {
        "status": prob.status,
        "x_star": x_star,
        "lambda_star": lambda_star,
        "v_star": v_star,
        "p_star": p_star
    }

# -----------------------------------------
# 6) Kiểm tra KKT bằng số (tổng quát)
# -----------------------------------------
def check_kkt_numeric(spec: KKTProblemSpec, x_star, lambda_star, v_star, tol=1e-6):
    A, b, c, s = spec.A, spec.b, spec.c, spec.s
    x_star = np.asarray(x_star, dtype=float).reshape(-1)
    lambda_star = np.asarray(lambda_star, dtype=float).reshape(-1)

    # g(x) <= 0
    g = spec.g_linear(x_star)
    # h(x) = 0
    h = spec.h_linear(x_star)

    primal_ok = (np.all(g <= tol) and abs(h) <= 1e-5 and np.all(x_star >= -tol))
    dual_ok = np.all(lambda_star >= -tol)

    comp = lambda_star * g
    comp_ok = np.all(np.abs(comp) <= 1e-5)

    # stationarity: ∇f + Jg^T λ + Jh^T v = 0
    eps = 1e-12
    grad_f = np.log(np.maximum(x_star, eps)) + 1.0               # (n,)
    station = grad_f + spec.Jg().T @ lambda_star + spec.Jh().T.flatten() * v_star
    station_ok = np.all(np.abs(station) <= 1e-5)

    return {
        "primal_ok": primal_ok,
        "dual_ok": dual_ok,
        "comp_ok": comp_ok,
        "station_ok": station_ok,
        "g(x*)": g,
        "h(x*)": h,
        "comp_residual": comp,
        "station_residual": station
    }

# ============================================================
# MAIN DEMO: ĐÚNG BÀI BẠN ĐƯA (có thể đổi input tùy ý)
# ============================================================
if __name__ == "__main__":
    # ---------- INPUT ----------
    A = np.array([[1., 0., 0.],
                  [0., 2., 0.],
                  [0., 0., 3.]])
    b = np.array([0.5, 1.0, 1.5])
    c = np.ones(3)      # đổi thành vector khác nếu muốn: np.array([1,2,1]) ...
    s = 1.5
    ineq_type = ">="    # đổi thành "<=" nếu Ax <= b

    spec = KKTProblemSpec(A=A, b=b, c=c, s=s, ineq_type=ineq_type)

    # (a) + (b): In Lagrangian và KKT cụ thể (tự động theo subject to)
    print_lagrangian_and_kkt(spec)

    # (c): Giải primal/dual, lấy x*, (λ*, v*), p*
    sol = solve_primal_dual(spec, solver=None, verbose=False)
    print("\n===== (c) NGHIỆM THỎA KKT (từ nghiệm tối ưu primal/dual) =====")
    print("Status:", sol["status"])
    print("x* =", sol["x_star"])
    print("lambda* =", sol["lambda_star"])
    print("v* =", sol["v_star"])
    print("p* = Σ x_i log x_i =", sol["p_star"])

    # Kiểm tra KKT số
    kkt = check_kkt_numeric(spec, sol["x_star"], sol["lambda_star"], sol["v_star"])
    print("\n===== KIỂM TRA KKT (NUMERIC) =====")
    print("Primal feasible:", kkt["primal_ok"])
    print("Dual feasible  :", kkt["dual_ok"])
    print("Complementary  :", kkt["comp_ok"])
    print("Stationarity   :", kkt["station_ok"])
    print("g(x*) =", kkt["g(x*)"])
    print("h(x*) =", kkt["h(x*)"])
```
