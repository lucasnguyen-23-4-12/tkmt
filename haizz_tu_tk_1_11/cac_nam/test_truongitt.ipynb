# KTT

```python
import sympy as sp

# Định nghĩa các biến
x1, x2, lambda1, lambda2 = sp.symbols('x1 x2 lambda1 lambda2')

# Hàm mục tiêu
f = x1**2 + x2**2

# Ràng buộc
constraint1 = x1 + x2 - 1  # x1 + x2 >= 1
constraint2 = x1 - 2  # x1 = 2

# Hàm Lagrange
lagrange = f + lambda1 * constraint1 + lambda2 * constraint2

# Tính đạo hàm của hàm Lagrange
lagrange_x1_derivative = sp.diff(lagrange, x1)
lagrange_x2_derivative = sp.diff(lagrange, x2)

# Điều kiện KKT
kkt_equations = [
    lagrange_x1_derivative,
    lagrange_x2_derivative,
    constraint1,
    constraint2
]

# Giải hệ phương trình KKT
solutions_kkt = sp.solve(kkt_equations, (x1, x2, lambda1, lambda2))

# In nghiệm từ KKT
print("Nghiệm KKT:")
if isinstance(solutions_kkt, dict):  # Trường hợp trả về một dictionary duy nhất
    print(f"x1 = {solutions_kkt[x1]}, x2 = {solutions_kkt[x2]}, lambda1 = {solutions_kkt[lambda1]}, lambda2 = {solutions_kkt[lambda2]}")
elif isinstance(solutions_kkt, list):  # Trường hợp trả về danh sách nghiệm
    for sol in solutions_kkt:
        print(f"x1 = {sol[x1]}, x2 = {sol[x2]}, lambda1 = {sol[lambda1]}, lambda2 = {sol[lambda2]}")
else:
    print("Không tìm được nghiệm phù hợp.")
```

# Dual function

```python
import sympy as sp

# Định nghĩa các biến
x1, x2, lambda1, lambda2 = sp.symbols('x1 x2 lambda1 lambda2')

# Hàm mục tiêu
f = x1**2 + x2**2

# Ràng buộc
constraint1 = x1 + x2 - 1  # x1 + x2 >= 1
constraint2 = x1 - 2  # x1 = 2

# Hàm Lagrange
lagrange = f + lambda1 * constraint1 + lambda2 * constraint2

# Tính đạo hàm của hàm Lagrange theo x1 và x2
lagrange_x1_derivative = sp.diff(lagrange, x1)
lagrange_x2_derivative = sp.diff(lagrange, x2)

# Giải phương trình đạo hàm để tìm giá trị x1, x2 theo lambda1, lambda2
solutions_dual = sp.solve([lagrange_x1_derivative, lagrange_x2_derivative], (x1, x2))

# Thay giá trị x1 và x2 vào hàm Lagrange để tính hàm đối ngẫu q(lambda1, lambda2)
if isinstance(solutions_dual, dict):
    lagrange_with_solutions = lagrange.subs(solutions_dual)
elif isinstance(solutions_dual, list):
    lagrange_with_solutions = lagrange.subs({x1: solutions_dual[0][0], x2: solutions_dual[0][1]})
else:
    lagrange_with_solutions = None

if lagrange_with_solutions:
    q = sp.simplify(lagrange_with_solutions)
    print("\nHàm đối ngẫu q(lambda1, lambda2):")
    print(q)

    # Tối đa hóa hàm đối ngẫu bằng cách tính đạo hàm theo lambda1 và lambda2
    lagrange_lambda1_derivative = sp.diff(q, lambda1)
    lagrange_lambda2_derivative = sp.diff(q, lambda2)

    # Giải hệ phương trình đạo hàm
    dual_solutions = sp.solve([lagrange_lambda1_derivative, lagrange_lambda2_derivative], (lambda1, lambda2))

    # In nghiệm của bài toán đối ngẫu
    print("\nNghiệm bài toán đối ngẫu:")
    if isinstance(dual_solutions, dict):
        print(f"lambda1 = {dual_solutions[lambda1]}, lambda2 = {dual_solutions[lambda2]}")
    elif isinstance(dual_solutions, list):
        for sol in dual_solutions:
            print(f"lambda1 = {sol[0]}, lambda2 = {sol[1]}")
    else:
        print("Không tìm được nghiệm đối ngẫu phù hợp.")

    # Tính giá trị tối ưu của hàm đối ngẫu tại nghiệm
    if isinstance(dual_solutions, dict):
        optimal_dual_value = q.subs(dual_solutions)
    elif isinstance(dual_solutions, list):
        optimal_dual_value = q.subs({lambda1: dual_solutions[0][0], lambda2: dual_solutions[0][1]})
    else:
        optimal_dual_value = None

    if optimal_dual_value is not None:
        print(f"\nGiá trị tối ưu của hàm đối ngẫu: {optimal_dual_value}")
    else:
        print("Không tính được giá trị tối ưu của hàm đối ngẫu.")
else:
    print("Không tính được hàm đối ngẫu.")
```
