```python
from scipy.optimize import linprog

# Hệ số của hàm mục tiêu (chúng ta sẽ đổi dấu để thành bài toán minimize)
c = [-250000, -180000]  # Maximizing so we negate the objective function

# Ma trận hệ số ràng buộc (A)
A = [
    [12, 9],  # Số linh kiện
    [1, 0],   # Giới hạn số radio kiểu 1
    [0, 1]    # Giới hạn số radio kiểu 2
]

# Vector giới hạn ràng buộc (b)
b = [900, 45, 80]

# Giới hạn (bound) cho các biến x1 và x2 (x1, x2 ≥ 0)
x1_bounds = (0, None)
x2_bounds = (0, None)

# Giải bài toán tối ưu
result = linprog(c, A_ub=A, b_ub=b, bounds=[x1_bounds, x2_bounds], method='highs')

# Kết quả
print("Số radio kiểu 1 cần sản xuất:", result.x[0])
print("Số radio kiểu 2 cần sản xuất:", result.x[1])
print("Lợi nhuận tối đa:", -result.fun)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm mục tiêu f(x)
def f(x1, x2):
    A = np.array([[1, 0], [0, -2]])
    b = np.array([2, 4])
    return np.exp(2*x1) + 0.5 * np.linalg.norm(A.dot([x1, x2]) - b)**2

# Gradient của f(x)
def grad_f(x1, x2):
    A = np.array([[1, 0], [0, -2]])
    b = np.array([2, 4])
    grad_x1 = 2 * np.exp(2 * x1) + (A[0, 0] * (A[0, 0] * x1 + A[0, 1] * x2 - b[0]))
    grad_x2 = (A[1, 1] * (A[1, 0] * x1 + A[1, 1] * x2 - b[1]))
    return np.array([grad_x1, grad_x2])

# Thuật toán Gradient Descent
def gradient_descent(learning_rate, x_init, max_iters=100):
    x_vals = [x_init]
    f_vals = [f(x_init[0], x_init[1])]
    
    x = x_init
    for _ in range(max_iters):
        grad = grad_f(x[0], x[1])
        x = x - learning_rate * grad
        x_vals.append(x)
        f_vals.append(f(x[0], x[1]))
    
    return np.array(x_vals), np.array(f_vals)

# Giá trị ban đầu
x_init = np.array([-3, 5])

# Thực hiện với learning rate lần lượt là 2/5 và 1/2
x_vals_1, f_vals_1 = gradient_descent(learning_rate=2/5, x_init=x_init)
x_vals_2, f_vals_2 = gradient_descent(learning_rate=1/2, x_init=x_init)

# Vẽ đồ thị hội tụ
plt.plot(range(len(f_vals_1)), f_vals_1 - min(f_vals_1), label='Learning Rate = 2/5')
plt.plot(range(len(f_vals_2)), f_vals_2 - min(f_vals_2), label='Learning Rate = 1/2')
plt.yscale('log')  # Sử dụng thang logarit cho độ sai số
plt.xlabel('Số lần lặp')
plt.ylabel('Sai số |f(x^k) - p*|')
plt.legend()
plt.show()
```

### Câu 3. (3 điểm, CLO 1, 2) ###
Cho các vector $c, a_1, a_2, \ldots, a_m \in \mathbb{R}^n$ và các số $b_1, b_2, \ldots, b_m \in \mathbb{R} \ (m \in \mathbb{Z}^+)$, xét bài toán tối ưu sau:
$$\min_{x \in \mathbb{R}^n} f(x) = c^T x - \sum_{i=1}^{m} \log(b_i - a_i^T x).$$

### 1. Tính $\nabla f$ và $\nabla^2 f$: ###
Với hàm mục tiêu:
$$
f(x) = c^T x - \sum_{i=1}^{m} \log(b_i - a_i^T x),
$$

Gradient của $f$ là:
$$
\nabla f(x) = c - \sum_{i=1}^{m} \frac{a_i}{b_i - a_i^T x}.
$$

Hessian của $f$ là:
$$
\nabla^2 f(x) = \sum_{i=1}^{m} \frac{a_i a_i^T}{(b_i - a_i^T x)^2}.
$$

### 2. Chứng minh $f$ là hàm lồi: ###
Để chứng minh $f(x)$ là hàm lồi, ta cần kiểm tra xem Hessian $\nabla^2 f(x)$ có xác định dương hay không. Nếu $\nabla^2 f(x)$ xác định dương, thì $f(x)$ là hàm lồi.

Vì $(b_i - a_i^T x) > 0$ với mọi $x$ trong miền khả thi, nên $\nabla^2 f(x)$ là tổng của các ma trận xác định dương $\frac{a_i a_i^T}{(b_i - a_i^T x)^2}$. Do đó, $\nabla^2 f(x)$ cũng xác định dương, và ta có thể kết luận rằng $f(x)$ là hàm lồi.

### 3. Xác định điểm tối ưu $x^*$ và giá trị tối ưu $p^*$: ###
Với các giá trị đã cho:
$$
A = \begin{bmatrix} 0 & -2 & -1 & -2 \\ -1 & -3 & 0 & 0 \\ 1 & -1 & 2 & -5 \end{bmatrix}, \quad b = (3, -1, 3, 2)^T, \quad c = \left( 4, \frac{11}{4}, \frac{13}{4} \right)^T.
$$

Từ hàm mục tiêu, ta tìm các điều kiện khả thi để giải bài toán tối ưu. Ta cần giải hệ phương trình:
$$
\nabla f(x) = 0.
$$



```python
#Bai 3 c 
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Ma trận A và các vector b, c
A = np.array([[0, -2, -2], [1, -1, -2], [1, -1, -5]])
b = np.array([3, -1, 2])
c = np.array([4, 11/4, 13/4])

# Hàm mục tiêu
def f(x):
    m = len(b)
    sum_log = 0
    for i in range(m):
        sum_log += np.log(b[i] - np.dot(A[i], x))
    return np.dot(c, x) - sum_log

# Gradient của hàm mục tiêu
def grad_f(x):
    m = len(b)
    grad = c.copy()
    for i in range(m):
        grad -= A[i] / (b[i] - np.dot(A[i], x))
    return grad

# Thuật toán Gradient Descent
def gradient_descent(x0, learning_rate, max_iter=100):
    x = x0
    x_vals = [x0]
    f_vals = [f(x0)]
    for i in range(max_iter):
        grad = grad_f(x)
        x = x - learning_rate * grad
        x_vals.append(x)
        f_vals.append(f(x))
    return np.array(x_vals), np.array(f_vals)

# Khởi tạo
x0 = np.array([-1, 1, 0])  # Giá trị khởi tạo
learning_rate = 0.01  # Tốc độ học
max_iter = 100  # Số vòng lặp

# Thực hiện Gradient Descent
x_vals, f_vals = gradient_descent(x0, learning_rate, max_iter)

# Vẽ đồ thị sai số
plt.plot(np.abs(f_vals - min(f_vals)))
plt.xlabel('Số lần lặp (Iteration)')
plt.ylabel('Sai số |f(x^{(k)}) - p^*|')
plt.yscale('log')  # Thang đo logarit cho sai số
plt.title('Sai số Gradient Descent')
plt.show()

# In ra kết quả tối ưu
print("Điểm tối ưu x*: ", x_vals[-1])
print("Giá trị tối ưu f(x*): ", f_vals[-1])
```
