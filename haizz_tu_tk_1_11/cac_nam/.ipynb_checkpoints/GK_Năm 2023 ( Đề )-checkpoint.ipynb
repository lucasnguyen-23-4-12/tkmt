# Câu 1: ( 3 điểm )
Giả sử nhu cầu tối thiểu mỗi ngày về các chất dinh dưỡng đạm, đường, khoáng cho một loại gia súc tương ứng là 90g, 130g, 10g. Cho biết hàm lượng các chất dinh dưỡng trên có trong 1g thức ăn A, B, C và giá mua 1kg thức ăn cho mỗi loại được cho trong bảng sau:

\begin{array}{|c|c|c|c|}
\hline
\text{Chất dinh dưỡng} & \text{A} & \text{B} & \text{C} \\
\hline
\text{Đạm} & 0.1 \, \text{kg} & 0.2 \, \text{kg} & 0.3 \, \text{kg} \\
\hline
\text{Đường} & 0.3 \, \text{kg} & 0.4 \, \text{kg} & 0.2 \, \text{kg} \\
\hline
\text{Khoáng} & 0.02 \, \text{kg} & 0.01 \, \text{kg} & 0.03 \, \text{kg} \\
\hline
\text{Giá mua (VND)} & 3000 & 4000 & 5000 \\
\hline
\end{array}

Bài toán đặt ra là xác định khối lượng thức ăn mỗi loại cần mua để tổng số tiền chi cho việc mua thức ăn là ít nhất nhưng vẫn đáp ứng được nhu cầu dinh dưỡng mỗi ngày.

1. Hãy lập mô hình tối ưu của bài toán trên theo dạng dưới đây, bằng cách chỉ ra hàm mục tiêu $f_0$, ma trận $A$ và vector $b$.

\begin{align}
\text{minimize} \quad f_0 (x), \\
\text{subject} \quad \text{to} \quad Ax \leq b.
\end{align}

2. Sử dụng scipy.optimize.lineprog để giải bài toán trên.
3. Sử dụng cvxpy để giải bài toán trên.

````python
import numpy as np
from scipy.optimize import linprog
import cvxpy as cp

# 1. Định nghĩa các hệ số của hàm mục tiêu
c = np.array([3000, 4000, 5000])

# 2. Định nghĩa các hệ số của các ràng buộc bất đẳng thức
A = np.array([
    [-0.1, -0.2, -0.3],  # Đạm
    [-0.3, -0.4, -0.2],  # Đường
    [-0.02, -0.01, -0.03]  # Khoáng
])
b = np.array([-90, -130, -10])

# 3. Định nghĩa các giới hạn cho các biến
x0_bounds = (0, None)
x1_bounds = (0, None)
x2_bounds = (0, None)

# Giải quyết vấn đề lập trình tuyến tính bằng scipy.optimize.lineprog
res_scipy = linprog(c, A_ub=A, b_ub=b, bounds=(x0_bounds, x1_bounds, x2_bounds))
print("Giá trị tối ưu (scipy):", res_scipy.fun)
print("Khối lượng thức ăn mỗi loại (scipy):", res_scipy.x)

# Giải quyết vấn đề lập trình tuyến tính bằng cvxpy
x = cp.Variable(3)
objective = cp.Minimize(c @ x)
constraints = [A @ x <= b, x >= 0]
prob = cp.Problem(objective, constraints)
prob.solve()
print("Giá trị tối ưu (cvxpy):", prob.value)
print("Khối lượng thức ăn mỗi loại (cvxpy):", x.value)
````

3. Hãy vẽ đồ thị biểu thị cho sai số $err_k$ = $|f(x^{(k)}) - p^\ast|$. Từ đó anh/chị rút ra được kết luận gì?

````python
import numpy as np

# Phần 1: Tìm giá trị tối ưu p* của bài toán (1)
def f(x):
    return 0.5 * x[0]**2 + 0.5 * x[1]**2 + 2*x[1]

def grad_f(x):
    return np.array([x[0], x[1] + 2])

x_opt = np.array([0, -2])  # Giá trị tối ưu
p_opt = f(x_opt)  # Giá trị tối ưu của hàm số

# Phần 2: Sử dụng thuật toán Gradient Descent cho bài toán (1) với learning rate được tính theo phương pháp exact line search
def gradient_descent(x0, max_iter=1000, tol=1e-6):
    x = x0
    x_history = [x]
    f_history = [f(x)]
    for _ in range(max_iter):
        grad = grad_f(x)
        alpha = 1 / (1 + np.linalg.norm(grad))  # Learning rate được tính theo phương pháp exact line search
        x = x - alpha * grad
        x_history.append(x)
        f_history.append(f(x))
        if np.linalg.norm(grad) < tol:
            break
    return x, x_history, f_history

x0 = np.array([1, 0])  # Điểm khởi tạo
x_opt_gd, x_history, f_history = gradient_descent(x0)

# Phần 3: Vẽ đồ thị biểu thị cho sai số err_k = |f(x^(k)) - p*|
import matplotlib.pyplot as plt

err_history = [abs(f_x - p_opt) for f_x in f_history]
plt.plot(err_history)
plt.xlabel('Số vòng lặp')
plt.ylabel('Sai số')
plt.title('Đồ thị biểu thị cho sai số err_k')
plt.show()

# Kết luận: Từ đồ thị biểu thị cho sai số err_k, ta có thể thấy rằng sau mỗi vòng lặp, sai số giảm dần và tiến gần đến 0. Điều này chứng tỏ rằng thuật toán Gradient Descent có thể tìm được giá trị tối ưu của hàm số.

````

Bước 1: Lập mô hình tối ưu
Hàm mục tiêu

Chúng ta sử dụng dấu âm vì scipy.optimize.linprog mặc định là bài toán tối thiểu hóa.

Ràng buộc bất đẳng thức
Các ràng buộc bất đẳng thức là:

\begin{align} 2x_1 + x_2 + 3x_3 &\leq 500 \quad \text{(ràng buộc lao động)} \ 100x_1 + 40x_2 + 250x_3 &\leq 40000 \quad \text{(ràng buộc chi phí sản xuất)} \end{align}

Ràng buộc đẳng thức
Ràng buộc đẳng thức là:

$ x_1 - 6x_2 = 0 \quad \text{(tỉ lệ số bàn và ghế)} $

````python
import numpy as np
from scipy.optimize import linprog

# Định nghĩa các hệ số trong bài toán
c = [-260, -120, -600]  # Hệ số âm vì scipy.optimize.linprog mặc định là bài toán tối thiểu hóa

# Ma trận A và vector b cho các ràng buộc bất đẳng thức Ax <= b
A = [
    [2, 1, 3],  # Ràng buộc lao động
    [100, 40, 250],  # Ràng buộc chi phí sản xuất
]
b = [500, 40000]  # Số lao động và chi phí sản xuất

# Ma trận A_eq và vector b_eq cho các ràng buộc đẳng thức A_eq x = b_eq
A_eq = [
    [1, -6, 0],  # Ràng buộc số bàn và ghế theo tỉ lệ 1/6
]
b_eq = [0]

# Giải bài toán bằng scipy.optimize.linprog
result_scipy = linprog(c, A_ub=A, b_ub=b, A_eq=A_eq, b_eq=b_eq, method='highs')

# Kết quả từ scipy.optimize.linprog
print("Kết quả từ scipy.optimize.linprog:")
print(f"Số lượng bàn: {result_scipy.x[0]}")
print(f"Số lượng ghế: {result_scipy.x[1]}")
print(f"Số lượng tủ: {result_scipy.x[2]}")
print(f"Tổng doanh thu: {-result_scipy.fun}")
````

````python
import cvxpy as cp

# Định nghĩa biến x
x = cp.Variable(3)

# Định nghĩa hàm mục tiêu
objective = cp.Maximize(c @ x)

# Định nghĩa các ràng buộc
constraints = [
    A @ x <= b,
    A_eq @ x == b_eq,
    x >= 0
]

# Giải bài toán bằng cvxpy
problem = cp.Problem(objective, constraints)
result_cvxpy = problem.solve()

# Kết quả từ cvxpy
print("\nKết quả từ cvxpy:")
print(f"Số lượng bàn: {x.value[0]}")
print(f"Số lượng ghế: {x.value[1]}")
print(f"Số lượng tủ: {x.value[2]}")
print(f"Tổng doanh thu: {result_cvxpy}")
````

Code mẫu (câu 2.1, 2.2):

Câu 2.1: Tính đạo hàm của hàm số
Cho hàm số mục tiêu:

$ f(x) = x^2 + \sin\left(\frac{x}{2}\right) + 2023x $

Đạo hàm của hàm số này là:

$ f'(x) = 2x + \frac{1}{2}\cos\left(\frac{x}{2}\right) + 2023 $

````python
import numpy as np

# 1. Tính đạo hàm của hàm số f(x)
def f_prime(x):
    return 2*x + (1/2)*np.cos(x/2) + 2023

# 2. Sử dụng thuật toán Gradient Descent để tối ưu hóa hàm số f(x)
def gradient_descent(x0, learning_rate, max_iterations):
    x = x0
    for i in range(max_iterations):
        grad = f_prime(x)
        x = x - learning_rate * grad
        fx = x**2 + np.sin(x/2) + 2023*x
        print(f"Iteration {i+1}: x = {x}, f(x) = {fx}")

# Giá trị ban đầu của x
x0 = 2024

# Learning rate
learning_rate = 0.9

# Số vòng lặp tối đa
max_iterations = 100

# Thực hiện thuật toán Gradient Descent
gradient_descent(x0, learning_rate, max_iterations)
````

Nếu learning rate quá lớn, thuật toán có thể bỏ qua điểm tối ưu và dao động xung quanh nó. Nếu learning rate quá nhỏ, thuật toán có thể hội tụ rất chậm và mất nhiều thời gian để đạt được điểm tối ưu.

Để chọn giá trị tối ưu của learning rate, bạn có thể thử nghiệm với các giá trị khác nhau và quan sát tốc độ hội tụ của thuật toán. Một cách tiếp cận phổ biến là sử dụng phương pháp giảm dần learning rate theo thời gian hoặc sử dụng các kỹ thuật như learning rate annealing.

Câu 2.4: Ví dụ thực tế
Một ví dụ cụ thể về một vấn đề thực tế trong học máy mà Gradient Descent có thể được sử dụng để giải quyết là huấn luyện mô hình hồi quy tuyến tính. - $\alpha = \gamma_7 = 7$
- $\beta = \sum_{i=1}^8 \gamma_i = 1 + 2 + 3 + 4 + 5 + 6 + 7 + 8 = 36$

Bài toán tối ưu là:

\[
\min_{x=(x_1,x_2)\in \mathbb{R}^2} f(x) = \frac{1}{2}x_1^2 + \frac{1}{2}x_2^2 + 8x_2 - 36
\]

Để tìm giá trị tối ưu $p^*$, ta cần tìm điểm $x^*$ sao cho $\nabla f(x^*) = 0$.

Đạo hàm của hàm mục tiêu $f(x)$ là:

\[
\nabla f(x) = \begin{bmatrix}
x_1 \\
x_2 + 8
\end{bmatrix}
\]

Giải hệ phương trình $\nabla f(x) = 0$:

\[
\begin{cases}
x_1 = 0 \\
x_2 + 8 = 0
\end{cases}
\]

Ta có $x^* = (0, -8)$.

Giá trị tối ưu $p^*$ là:

\[
p^* = f(x^*) = \frac{1}{2}(0)^2 + \frac{1}{2}(-8)^2 + 8(-8) - 36 = -68
\]

### Câu 3.2: Sử dụng thuật toán Gradient Descent

Với learning rate $\eta = \frac{1}{\alpha + 1} = \frac{1}{8}$ và điểm khởi tạo $x^{(0)} = (1, 0)$, công thức cập nhật điểm $x^{(k)}$ là:

\[
x^{(k+1)} = x^{(k)} - \eta \nabla f(x^{(k)})
\]

### Câu 3.3: Vẽ đồ thị biểu thị sai số

Dưới đây là đoạn code Python để thực hiện các bước trên và vẽ đồ thị biểu thị sai số:

```python
import numpy as np
import matplotlib.pyplot as plt

# Giả sử mã số sinh viên là 12345678
gamma = [1, 2, 3, 4, 5, 6, 7, 8]
alpha = gamma[6]
beta = sum(gamma)

# Bài toán tối ưu
def f(x):
    return 0.5 * x[0]**2 + 0.5 * x[1]**2 + (alpha + 1) * x[1] - beta

# Đạo hàm của hàm mục tiêu
def grad_f(x):
    return np.array([x[0], x[1] + (alpha + 1)]) 

# Tìm giá trị tối ưu p*
x_star = np.array([0, - (alpha + 1)])
p_star = f(x_star)

print(f"alpha = {alpha}")
print(f"beta = {beta}")
print(f"Giá trị tối ưu p* = {p_star}")

# Thuật toán Gradient Descent
def gradient_descent(x0, learning_rate, max_iterations):
    x = x0
    x_values = [x]
    for _ in range(max_iterations):
        grad = grad_f(x)
        x = x - learning_rate * grad
        x_values.append(x)
    return x_values

# Giá trị khởi tạo và learning rate
x0 = np.array([1, 0])
learning_rate = 1 / (alpha + 1)
max_iterations = 100

# Thực hiện Gradient Descent
x_values = gradient_descent(x0, learning_rate, max_iterations)

# Tính sai số err_k
errors = [np.linalg.norm(grad_f(x_star) - grad_f(x))**2 for x in x_values]

# Vẽ đồ thị biểu thị sai số err_k
plt.plot(errors)
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error vs Iteration')
plt.show()
```

### Câu 3.4: Tìm learning rate theo phương pháp exact line search

Để tìm learning rate theo phương pháp exact line search, ta cần tìm $\eta$ sao cho:

\[
\eta = \frac{\nabla f(x)^T \nabla f(x)}{\nabla f(x)^T H \nabla f(x)}
\]

Trong đó $H$ là ma trận Hessian của $f(x)$. Với hàm $f(x)$ đã cho, ma trận Hessian là:

\[
H = \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
\]

Do đó, $\eta$ có thể được tính như sau:

```python
# Tìm learning rate theo phương pháp exact line search
def exact_line_search(x):
    grad = grad_f(x)
    return np.dot(grad, grad) / np.dot(grad, grad_f(grad))

learning_rates = [exact_line_search(x) for x in x_values]

print(f"Learning rate theo phương pháp exact line search: {learning_rates}")
```
