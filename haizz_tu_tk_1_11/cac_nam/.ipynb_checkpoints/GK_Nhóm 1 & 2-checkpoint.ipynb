## Câu 1 (3 điểm, CLO 1, 2)

Một công ty điện tử sản xuất hai kiểu radio trên hai dây chuyền độc lập. Radio kiểu một sản xuất trên dây chuyền một với công suất 45 radio/ngày, radio kiểu hai sản xuất trên dây chuyền hai với công suất 80 radio/ngày. Để sản xuất một chiếc radio kiểu một cần 12 linh kiện, để sản xuất một chiếc radio kiểu hai cần 9 linh kiện. Tiền lãi khi bán một chiếc radio kiểu một là 250.000 đồng, lãi thu được khi bán một chiếc radio kiểu hai là 180.000 đồng. Biết rằng số linh kiện có thể sử dụng tối đa trong một ngày là 900. Bài toán đặt ra là xác định lợi nhuận có thể thu được lớn nhất của công ty là bao nhiêu.

a) Hãy lập mô hình tối ưu của bài toán trên theo dạng dưới đây bằng cách chỉ ra hàm mục tiêu $f_0$, ma trận $A$ và vector $b$:
$$\begin{align}
&\text{minimize } f_0(x) \\
&\text{subject to } Ax \leq b
\end{align}$$
b) Sử dụng thư viện thích hợp để giải bài toán trên.

## Câu 2 (4 điểm, CLO 1, 2)

Cho ma trận $A$ và vector $b$ như sau:
$A = \begin{pmatrix}
1 & 0 \\
0 & -2
\end{pmatrix}$
$b = \begin{pmatrix}
2 ; 4
\end{pmatrix}$

Xét bài toán tối ưu:

$$\begin{align}
\min_{x = (x_1, x_2) \in \mathbb{R}^2} f(x) =  e^{2x_1} + \frac{1}{2} ||Ax - b||_2^2     (1)
\end{align}$$

a) ( 2 điểm ) Xác định điểm tối ưu $x^\ast$ và giá trị tối ưu $p^\ast$ của bài toán (1)

b) Sử dụng thuật toán Gradient Descent, với giá trị $x$ ban đầu là:
$x^{(0)} = \begin{pmatrix}
-3 ; 5
\end{pmatrix}$
sử dụng learning rate lần lượt là $\frac{2}{5}$ và $\frac{1}{2}$ và thực hiện tối đa 100 vòng lặp. In ra giá trị của $x^{(k)}$, $f(x^{(k)})$ tương ứng sau mỗi vòng lặp $k$ và vẽ đồ thị biểu thị cho sai số $|f(x^{(k)}) - p^\ast|$ trong hai trường hợp của learning rate. Hãy đưa ra kết luận về sự hội tụ của thuật toán Gradient Descent trong từng trường hợp của learning rate.

## Câu 3 (3 điểm, CLO 1,2)
Cho các vector $ c, a_1, a_2,...,a_m \in \mathbb{R}^n $ và các số $ b_1, b_2,...,b_m \in \mathbb{R}^n $ (với $ m \in \mathbb{Z}^+ $), xét bài toán tối ưu:
$$
\min_{x \in \mathbb{R}^n} f(x) = c^T x + \sum_{i=1}^{m} \log(b_i - a_i^T x)   (2)
$$
a) Tính $(\nabla f)$ và $(\nabla^2 f)$ và chứng minh hàm f là hàm lồi trên tập xác định của nó.

b) Cho $m = 4$, $n = 3$ và ma trận $A \in M_{3\times4}(\mathbb{R})$ được xác định như sau:
$$
A = \begin{bmatrix}
0 & -2 & -1 & -2 \\
-1 & 1 & -3 & 0 \\
1 & -1 & 2 & -5
\end{bmatrix}
$$
Gọi $a_i$ là cột thứ $i$ của $A$ $(i = 1,2,3,4)$,
$b = (3, -1, 3, 2)^T$ và
$c = (4, \frac{11}{4}, \frac{13}{4})$

Xác định điểm tối ưu $x^*$ và giá trị tối ưu $p^\ast$ của bài toán (2).

c) Với dữ kiện được cho ở ý b), sử dụng thuật toán Gradient Descent cho bài toán (2) với learning rate được tính theo phương pháp exact line search và điểm khởi tạo $x^{(0)} = (-1, 1, 0)^T$, vẽ đồ thị biểu thị cho sai số $err_k = |f(x^{(k)}) - p^\ast|$. Từ đó anh/chị rút ra được kết luận gì?
