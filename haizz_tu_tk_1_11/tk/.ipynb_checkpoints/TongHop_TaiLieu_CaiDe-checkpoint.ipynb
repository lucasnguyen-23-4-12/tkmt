a) Lập mô hình tối ưu:

Gọi $x_1$ là số xe Lead cần nhập
Gọi $x_2$ là số xe Vision cần nhập
1. Hàm mục tiêu (cần tối đa hóa lợi nhuận):
Lợi nhuận từ xe Lead: $5x_1$ (triệu đồng)
Lợi nhuận từ xe Vision: $3.2x_2$ (triệu đồng)
Do bài toán yêu cầu dạng minimize nên ta đổi dấu:
$f_0(x) = -(5x_1 + 3.2x_2)$
2. Các ràng buộc:
 1. Ràng buộc về vốn: $40x_1 + 30x_2 \leq 36000$
 2. Ràng buộc về tổng nhu cầu: $x_1 + x_2 \leq 1100$
 3. Ràng buộc về tỷ lệ nhu cầu: $x_1 \leq 1.5x_2$ hay $x_1 - 1.5x_2 \leq 0$
 4. Ràng buộc không âm: $x_1 \geq 0, x_2 \geq 0$

$$
\text{minimize } -5x_1 - 3.2x_2

$$
\begin{aligned}
40x_1 + 30x_2 &\leq 36000 \\
x_1 + x_2 &\leq 1100 \\
x_1 &\leq 1.5x_2 \\
x_1, x_2 &\geq 0
\end{aligned}
$$

### Câu 2:  Cho ma trận $A$ và vector $b$ như sau:
$A = \begin{pmatrix}
1 & 0 \\
0 & -2
\end{pmatrix}$
$b = \begin{pmatrix}
-\frac{12}{5} ,
-2024
\end{pmatrix}$

Xét bài toán tối ưu:
$\min_{x = (x_1, x_2) \in \mathbb{R}^2} f(x) = \frac{1}{2} \|Ax - b\|_2^2 - \log(1+x_1^2)$

a) Xác định điểm tối ưu $x^\ast$ và giá trị tối ưu $p^\ast$

b) Sử dụng thuật toán Gradient Descent, với giá trị $x$ ban đầu là:

$x^{(0)} = \begin{pmatrix}
-\frac{23}{10} \\
2024
\end{pmatrix}$
sử dụng learning rate lần lượt là $\frac{3}{10}$ và $\frac{3}{5}$ và thực hiện tối đa 100 vòng lặp. In ra giá trị của $x^{(k)}$, $f(x^{(k)})$ tương ứng sau mỗi vòng lặp $k$ và vẽ đồ thị biểu thị cho sai số $|f(x^{(k)}) - p^\ast|$ trong hai trường hợp của learning rate. Hãy đưa ra kết luận về sự hội tụ của thuật toán Gradient Descent trong từng trường hợp của learning rate.

```python
# a) Xác định điểm tối ưu x* và giá trị tối ưu y*
import numpy as np
A = np.array([[1, 0], [0, -2]])
b = np.array([-12/ 5, -2024])
A, b
```

```python
# Xác định điểm tối ưu x* và giá trị tối ưu y*
from scipy.optimize import minimize

# Define the function f(x)
def f(x):
    x1, x2 = x
    return 0.5 * np.linalg.norm(A @ x - b)**2 - np.log(1 + x1**2)

# Initial guess
x0 = np.array([-23/10, 2024])

# Use scipy.optimize.minimize to find the optimal x
result = minimize(f, x0, method='BFGS')

optimal_x = result.x
optimal_f = result.fun

print("Điểm tối ưu x*:", optimal_x)
print("Giá trị tối ưu p*:", optimal_f)
```

```python
# b) Sử dụng thuật toán Gradient Descent
import matplotlib.pyplot as plt
# Khởi tạo
# Initial guess
x0 = np.array([-23/10, 2024])

# Perform gradient descent for both learning rates
learning_rate_1 = 3/10
learning_rate_2 = 3/5
num_iters = 100

# Define the matrix A and vector b
A = np.array([[1, 0], [0, -2]])
b = np.array([-12/5, -2024])

# Define the objective function f(x)
def f(x):
    x1, x2 = x
    return 0.5 * np.linalg.norm(A @ x - b)**2 - np.log(1 + x1**2)

# Define the gradient of f(x)
def grad_f(x):
    x1, x2 = x
    grad_x1 = (A[0, 0] * (A[0, 0] * x1 - b[0])) - (2 * x1 / (1 + x1**2))
    grad_x2 = A[1, 1] * (A[1, 1] * x2 - b[1])
    return np.array([grad_x1, grad_x2])

# Gradient Descent function
def gradient_descent(x_init, learning_rate, num_iters=100):
    x = x_init
    history_x = []
    history_f = []
    
    for _ in range(num_iters):
        history_x.append(x)
        history_f.append(f(x))
        x = x - learning_rate * grad_f(x)
    
    return np.array(history_x), np.array(history_f)


# Gradient descent with learning rate 3/10
history_x_lr1, history_f_lr1 = gradient_descent(x0, learning_rate_1, num_iters)

# Gradient descent with learning rate 3/5
history_x_lr2, history_f_lr2 = gradient_descent(x0, learning_rate_2, num_iters)

# Vẽ đồ thị sai số |f(x^k) - f(x*)| theo số vòng lặp
plt.figure(figsize=(10, 6))
plt.plot(np.abs(history_f_lr1 - np.min(history_f_lr1)), label=f'Learning rate = {learning_rate_1}', color='b')
plt.plot(np.abs(history_f_lr2 - np.min(history_f_lr2)), label=f'Learning rate = {learning_rate_2}', color='r')
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('|f(x^(k)) - p*|')
plt.title('Convergence of Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()

# Output final values for inspection
print("Learning rate 3/10 - Final x and f(x):", history_x_lr1[-1], history_f_lr1[-1])
print("Learning rate 3/5 - Final x and f(x):", history_x_lr2[-1], history_f_lr2[-1])
```

### Câu 3:
Cho các vector $ c, a_1, a_2,...,a_m \in \mathbb{R}^n $ và các số $ b_1, b_2,...,b_m \in \mathbb{R}^n $ (với $ m \in \mathbb{Z}^+ $), xét bài toán tối ưu:

$$
\min_{x \in \mathbb{R}^n} f(x) = c^T x + \sum_{i=1}^{m} e^{(b_i - a_i^T x)}
$$

#### a) Tính Gradient ∇f và Hessian ∇²f, đồng thời chứng minh rằng f là hàm lồi trên tập xác định của nó.

1. **Gradient ∇f:**

Ta có hàm mục tiêu:

$$
f(x) = c^T x + \sum_{i=1}^{m} e^{(b_i - a_i^T x)}
$$

Gradient của $ f(x) $ theo vector $ x $ là:

$$
\nabla f(x) = c - \sum_{i=1}^{m} a_i e^{(b_i - a_i^T x)}
$$

Giải thích: Đạo hàm của $ e^{(b_i - a_i^T x)} $ theo $ x $ là:

$$
\nabla_x e^{(b_i - a_i^T x)} = - a_i e^{(b_i - a_i^T x)}
$$

2. **Hessian ∇²f:**

Hessian của $ f(x) $ là ma trận đạo hàm bậc hai của $ f(x) $ theo $ x $:

$$
\nabla^2 f(x) = \sum_{i=1}^{m} a_i a_i^T e^{(b_i - a_i^T x)}
$$

Giải thích: Đạo hàm bậc hai của $ - a_i e^{(b_i - a_i^T x)} $ là:

$$
\nabla_x^2 \left( - a_i e^{(b_i - a_i^T x)} \right) = a_i a_i^T e^{(b_i - a_i^T x)}
$$

3. **Chứng minh tính lồi của $ f(x) $:**

Một hàm số $ f(x) $ là hàm lồi nếu ma trận Hessian của nó là nửa xác định dương, tức là:

$$
\nabla^2 f(x) \geq 0
$$

Hessian của hàm $ f(x) $ là tổng của các ma trận có dạng:

$$
a_i a_i^T e^{(b_i - a_i^T x)}
$$

Trong đó:
- $ a_i a_i^T $ là ma trận nửa xác định dương (vì đó là tích của một vector và chuyển vị của chính nó),
- $ e^{(b_i - a_i^T x)} > 0 $ với mọi $ x $.

Do đó, Hessian $ \nabla^2 f(x) $ là tổng của các ma trận nửa xác định dương, nên $ f(x) $ là hàm lồi trên tập xác định của nó.



Tôi sẽ liệt kê các công thức đạo hàm thường gặp trong tối ưu:

### 1. Các công thức đạo hàm cơ bản:

#### Hàm mũ và logarit:
- $\frac{d}{dx}(e^x) = e^x$
- $\frac{d}{dx}(e^{ax}) = ae^{ax}$
- $\frac{d}{dx}(\ln x) = \frac{1}{x}$
- $\frac{d}{dx}(\log_a x) = \frac{1}{x\ln a}$

#### Hàm lũy thừa:
- $\frac{d}{dx}(x^n) = nx^{n-1}$
- $\frac{d}{dx}(x^{-n}) = -nx^{-n-1}$
- $\frac{d}{dx}(x^{\alpha}) = \alpha x^{\alpha-1}$

### 2. Đạo hàm của các hàm vector và ma trận:

#### Đạo hàm theo vector:
- $\nabla_x(a^Tx) = a$
- $\nabla_x(x^Ta) = a$
- $\nabla_x(x^TAx) = (A + A^T)x$
- $\nabla_x(a^TXb) = ab^T$

#### Đạo hàm bậc hai (Hessian):
- $\nabla^2_x(a^Tx) = 0$
- $\nabla^2_x(x^TAx) = A + A^T$
- $\nabla^2_x(\|Ax - b\|^2) = 2A^TA$

### 3. Đạo hàm của các hàm phức hợp:

#### Hàm norm:
- $\nabla_x(\|x\|_2) = \frac{x}{\|x\|_2}$
- $\nabla_x(\|x\|_2^2) = 2x$
- $\nabla_x(\|Ax - b\|_2^2) = 2A^T(Ax - b)$

#### Hàm lồi phổ biến:
- $\nabla_x(\log(\sum_{i=1}^n e^{x_i})) = \frac{e^{x_i}}{\sum_{j=1}^n e^{x_j}}$
- $\nabla_x(-\sum_{i=1}^n \log(x_i)) = -\frac{1}{x_i}$

### 4. Các công thức đặc biệt:

#### Đạo hàm của hàm tổng:
- $\nabla_x(\sum_{i=1}^m f_i(x)) = \sum_{i=1}^m \nabla f_i(x)$

#### Đạo hàm của hàm tích:
- $\nabla_x(f(x)g(x)) = f(x)\nabla g(x) + g(x)\nabla f(x)$

#### Đạo hàm của hàm hợp:
- $\nabla_x(f(g(x))) = f'(g(x))\nabla g(x)$

### 5. Các công thức trong tối ưu hóa:

#### Gradient của hàm mục tiêu phổ biến:
- $\nabla_x(c^Tx + \sum_{i=1}^m e^{a_i^Tx}) = c + \sum_{i=1}^m a_ie^{a_i^Tx}$
- $\nabla_x(\frac{1}{2}\|Ax - b\|_2^2) = A^T(Ax - b)$
- $\nabla_x(-\sum_{i=1}^m \log(b_i - a_i^Tx)) = \sum_{i=1}^m \frac{a_i}{b_i - a_i^Tx}$

#### Hessian của các hàm trên:
- $\nabla^2_x(c^Tx + \sum_{i=1}^m e^{a_i^Tx}) = \sum_{i=1}^m a_ia_i^Te^{a_i^Tx}$
- $\nabla^2_x(\frac{1}{2}\|Ax - b\|_2^2) = A^TA$
- $\nabla^2_x(-\sum_{i=1}^m \log(b_i - a_i^Tx)) = \sum_{i=1}^m \frac{a_ia_i^T}{(b_i - a_i^Tx)^2}$

## b) Cho $m = 4$, $n = 3$ và ma trận $A \in M_{3\times4}(\mathbb{R})$ được xác định như sau:
$$
A = \begin{bmatrix}
0 & -1 & 2 & \frac{1}{3} \\
-2 & 4 & -3 & -\frac{2}{7} \\
0 & -2 & 5 & 0
\end{bmatrix}
$$
Gọi $a_i$ là cột thứ $i$ của $A$ $(i = 1,2,3,4)$,
$b = (0, -1, 2, 2)^T$ và
$c = (-1+\frac{7e}{3}, 2-\frac{23e}{7}, -2+5e)^T$

Xác định điểm tối ưu $x^*$ của bài toán (2).

```python
import numpy as np
from scipy.optimize import fsolve

# Dữ liệu đề bài
A = np.array([
    [0, -1, 2, 1/3],
    [-2, 4, -3, -2/7],
    [0, -2, 5, 0]
])

b = np.array([0, -1, 2, 2])
c = np.array([-1 + 7 * np.exp(1) / 3, 2 - 23 * np.exp(1) / 7, -2 + 5 * np.exp(1)])

# Hàm gradient của f(x)
def gradient_f(x):
    grad = c.copy()
    for i in range(4):
        alpha_i = A[:, i]
        denom = np.exp(b[i] - np.dot(alpha_i, x))  # Tính e^(b_i - a_i^T x)
        grad -= alpha_i * denom  # Công thức gradient
    return grad

# Hàm tìm điểm tối ưu x*
def find_x_star(x0):
    return fsolve(gradient_f, x0)

# Hàm mục tiêu f(x)
def f(x):
    total = np.dot(c, x)
    for i in range(4):
        alpha_i = A[:, i]
        total += np.exp(b[i] - np.dot(alpha_i, x))  # Công thức f(x)
    return total

# Điểm khởi tạo
x0 = np.array([2, 0, 1])

# Tìm điểm tối ưu x* và giá trị tối ưu p*
x_star = find_x_star(x0)
p_star = f(x_star)

# Kết quả
print(f"Điểm tối ưu x*: {x_star}")
print(f"Giá trị tối ưu p*: {p_star}")
```

## c) Với dữ kiện được cho ở ý b), sử dụng thuật toán Gradient Descent cho bài toán (2) với learning rate được tính theo phương pháp exact line search và điểm khởi tạo $x^{(0)} = (2, 0, 1)^T$, vẽ đồ thị biểu thị cho sai số $err_k = |f(x^{(k)}) - f(x^)|$. Từ đó anh/chị rút ra được kết luận gì?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Hàm mục tiêu f(x)
def f(x, c, A, b):
    Ax = A.T @ x
    return c.T @ x + np.sum(np.exp(b - Ax))  # Sửa lại theo đề bài mới

# Gradient của f(x)
def grad_f(x, c, A, b):
    Ax = A.T @ x
    return c - A @ np.exp(b - Ax)  # Đạo hàm theo x với đề bài mới

# Tìm alpha với exact line search
def exact_line_search(x, d, c, A, b):
    func = lambda alpha: f(x + alpha * d, c, A, b)
    res = minimize_scalar(func)
    return res.x

# Thuật toán Gradient Descent với exact line search
def gradient_descent(x0, c, A, b, max_iter=100, tol=1e-6):
    x = x0
    history = []
    for k in range(max_iter):
        grad = grad_f(x, c, A, b)
        if np.linalg.norm(grad) < tol:
            print(f"Convergence reached at iteration {k+1}")
            break
        alpha = exact_line_search(x, -grad, c, A, b)
        x = x - alpha * grad
        fx = f(x, c, A, b)
        history.append((x.copy(), fx))
        print(f"Iteration {k+1}: x = {x}, f(x) = {fx}, alpha = {alpha}")
    return history

# Dữ liệu được cho trong đề bài
A = np.array([
    [0, -1, 2, 1/3],
    [-2, 4, -3, -2/7],
    [0, -2, 5, 0]
])
b = np.array([0, -1, 2, 2])
c = np.array([-1 + 7 * np.exp(1) / 3, 2 - 23 * np.exp(1) / 7, -2 + 5 * np.exp(1)])

# Điểm khởi tạo
x0 = np.array([2, 0, 1])

# Thực hiện gradient descent
history = gradient_descent(x0, c, A, b)

# Trích xuất các giá trị hàm f và lỗi sai số
f_values = [fx for _, fx in history]
p_star = min(f_values) if f_values else np.inf
errors = [abs(fx - p_star) for fx in f_values]

# Vẽ đồ thị sai số
plt.figure(figsize=(12, 6))
plt.plot(errors, label='Error |f(x^k) - p^*|')
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.legend()
plt.title('Convergence of Gradient Descent with Exact Line Search')
plt.show()
```

#### 2. Sử dụng thuật toán Gradient Descent cho bài toán (1) với learning rate được tính theo phương pháp exact line search và điểm khởi tạo $x^{(0)}$ = (1, 0), hãy tìm công thức cho điểm cập nhật $x^{(k)}$.

```python
import numpy as np

# Phần 1: Tìm giá trị tối ưu p* của bài toán (1)
def f(x):
    return 0.5 * x[0]**2 + 0.5 * x[1]**2 + 2*x[1]

def grad_f(x):
    return np.array([x[0], x[1] + 2])

x_opt = x_opttam  # Giá trị tối ưu
p_opt = f(x_opt)  # Giá trị tối ưu của hàm số

# Phần 2: Sử dụng thuật toán Gradient Descent cho bài toán (1) với learning rate được tính theo phương pháp exact line search
def gradient_descent(x0, max_iter=1000, tol=1e-6):
    x = x0
    x_history = [x]
    f_history = [f(x)]
    for i in range(max_iter):
        grad = grad_f(x)
        alpha = 1 / (1 + np.linalg.norm(grad))  # Learning rate được tính theo phương pháp exact line search
        x = x - alpha * grad
        x_history.append(x)
        f_history.append(f(x))
        # 🟢 In thông tin mỗi vòng lặp
        print(f"Vòng lặp {i+1}:")
        print(f"   Gradient = {grad.round()}")
        print(f"   Learning rate (alpha) = {alpha.round()}")
        print(f"   x = {x.round()}")
        print(f"   f(x) = {f(x)}")
        print("-"*40)
        # Điều kiện dừng
        if np.linalg.norm(grad) < tol:
            break
    return x, x_history, f_history

x0 = np.array([1, 0])  # Điểm khởi tạo
x_opt_gd, x_history, f_history = gradient_descent(x0)
```

### 3. Bằng cách thay đổi giá trị learning rate trong thuật toán trên, hãy giải thích sự quan trọng của việc điều chỉnh tham số learning rate trong thuật toán Gradient Descent và cách lựa chọn giá trị tối ưu của nó.


```python
# 3. Bằng cách thay đổi giá trị learning rate trong thuật toán trên, 
# hãy giải thích sự quan trọng của việc điều chỉnh tham số learning rate trong thuật toán Gradient Descent và cách lựa chọn giá trị tối ưu của nó.

import numpy as np
import matplotlib.pyplot as plt

# Hàm mục tiêu
def f(x):
    return x**2 + np.sin(x/2) + 2023*x

# Đạo hàm (gradient)
def grad_f(x):
    return 2*x + 0.5*np.cos(x/2) + 2023

# Hàm mô phỏng Gradient Descent
def gradient_descent(x0, lr, max_iter=50):
    x = x0
    history = [x]
    for _ in range(max_iter):
        x -= lr * grad_f(x)
        history.append(x)
    return np.array(history)

# Thử nhiều learning rate khác nhau
etas = [0.01, 0.1, 0.5, 0.9]
colors = ['blue', 'green', 'orange', 'red']

x0 = 2024
plt.figure(figsize=(8,5))
for eta, c in zip(etas, colors):
    x_hist = gradient_descent(x0, eta)
    plt.plot(x_hist, label=f"lr={eta}", color=c)

plt.xlabel('Số vòng lặp (iteration)')
plt.ylabel('Giá trị x')
plt.title('Ảnh hưởng của learning rate trong Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Việc điều chỉnh learning rate là cực kỳ quan trọng vì nó quyết định tốc độ và độ ổn định hội tụ.
# Qua mô phỏng, ta thấy learning rate quá nhỏ làm hội tụ chậm, còn quá lớn khiến thuật toán dao động hoặc phân kỳ.
# Giá trị η vừa phải (khoảng 0.1–0.4 với hàm này) cho hội tụ nhanh và ổn định nhất.
```

Câu 2.3: Tầm quan trọng của learning rate Learning rate là một tham số quan trọng trong thuật toán Gradient Descent vì nó quyết định bước nhảy của mỗi lần cập nhật giá trị ( x ). Nếu learning rate quá lớn, thuật toán có thể bỏ qua điểm tối ưu và dao động xung quanh nó. Nếu learning rate quá nhỏ, thuật toán có thể hội tụ rất chậm và mất nhiều thời gian để đạt được điểm tối ưu.

#### 4. Hãy tìm learning rate trong thuật toán Gradient Descent của bài toán (1) theo phương pháp exact line search. Từ đó rút ra nhận xét cho các điểm cập nhật $x^{(k)}$.

```python
import numpy as np

alpha = 9
beta = 28

A = np.eye(2)

def f(x):
    x1, x2 = x
    return 0.5*x1**2 + 0.5*x2**2 + (alpha+1)*x2 - beta

def grad_f(x):
    x1, x2 = x
    return np.array([x1, x2 + (alpha+1)])

def exact_line_search(grad):
    if np.allclose(grad, 0):  # nếu gradient gần 0 thì trả về 0 để tránh NaN
        return 0
    return (grad.T @ grad) / (grad.T @ (A @ grad))

x_0 = np.array([1.0, 0.0])
max_iter = 10

x_vals_exact = [x_0]
f_vals_exact = [f(x_0)]
learning_rates = []

x = x_0.copy()

print("===== PHẦN 3: EXACT LINE SEARCH =====")
for k in range(max_iter):
    grad = grad_f(x)
    lr_exact = exact_line_search(grad)
    learning_rates.append(lr_exact)
    x = x - lr_exact * grad
    x_vals_exact.append(x)
    f_vals_exact.append(f(x))
    print(f"Vòng lặp {k}: x = {x}, f(x) = {f(x)}, lr = {lr_exact}")
    if np.linalg.norm(grad) < 1e-10:
        print(f"✅ Hội tụ tại vòng {k}")
        break

print("\n===== KẾT QUẢ CUỐI CÙNG =====")
print(f"x* = {x_vals_exact[-1]}")
print(f"f(x*) = {f_vals_exact[-1]}")
```

Với Exact Line Search, tại vòng đầu tiên learning rate = 1 giúp thuật toán đi đúng một bước đến điểm cực tiểu x∗=[0,−10].

```python
import matplotlib.pyplot as plt

# Định nghĩa hàm Gradient
def gradient(x):
    A = np.array([[1, 0], [0, -2]])
    b = np.array([2, 4])
    grad = np.zeros(2)
    grad[0] = 2*np.exp(2*x[0]) + (x[0] - 2)
    grad[1] = 4*x[1] + 8
    return grad

# Hàm Gradient Descent
def gradient_descent(initial_x, learning_rate, max_iter):
    x = initial_x.copy()
    p_star_values = []  # Lưu giá trị f(x_k)
    errors = []  # Lưu sai số |f(x_k) - p^*|
    
    for k in range(max_iter):
        # --- Cập nhật trước (x^(k+1)) ---
        x = x - learning_rate * gradient(x)

        # --- Sau đó tính f(x^(k+1)) ---
        f_k = f(x)
        error = abs(f_k - p_star)

        # --- Lưu và in ra ---
        p_star_values.append(f_k)
        errors.append(error)
        print(f"Vòng lặp {k+1:3d}: x = {x.round(6)}, f(x) = {f_k:.6f}, sai số = {error:.6f}")

    return x, p_star_values, errors
# Giá trị ban đầu và tham số
initial_x = np.array([-3, 5])
max_iter = 100

# Learning rate 1: 2/5
learning_rate1 = 2 / 5
print('Learning rate 1: 2/5\n')
final_x1, p_star_values1, errors1 = gradient_descent(initial_x.copy(), learning_rate1, max_iter)

# Learning rate 2: 1/2
learning_rate2 = 1 / 2
print('Learning rate 2: 1/2\n')
final_x2, p_star_values2, errors2 = gradient_descent(initial_x.copy(), learning_rate2, max_iter)
```

```python
import numpy as np
from scipy.optimize import fsolve, minimize_scalar
import matplotlib.pyplot as plt

# ============================================
# DỮ LIỆU ĐỀ BÀI
# ============================================
A = np.array([
    [0, -1, 2, 1/3],
    [-2, 4, -3, -2/7],
    [0, -2, 5, 0]
])
b = np.array([0, -1, 2, 2])
c = np.array([
    -1 + 7 * np.exp(1) / 3,
    2 - 23 * np.exp(1) / 7,
    -2 + 5 * np.exp(1)
])

# ============================================
# ĐỊNH NGHĨA HÀM MỤC TIÊU VÀ ĐẠO HÀM
# ============================================
def f(x, c, A, b):
    Ax = A.T @ x
    return c.T @ x + np.sum(np.exp(b - Ax))

def grad_f(x, c, A, b):
    Ax = A.T @ x
    return c - A @ np.exp(b - Ax)

# ============================================
# (b) TÌM ĐIỂM TỐI ƯU x* BẰNG fsolve
# ============================================
def find_x_star(x0, c, A, b):
    func = lambda x: grad_f(x, c, A, b)
    return fsolve(func, x0)

x0 = np.array([2., 0., 1.])
x_star = find_x_star(x0, c, A, b)
p_star = f(x_star, c, A, b)

print("===== KẾT QUẢ CÂU (b) =====")
print(f"x* = {x_star}")
print(f"p* = {p_star:.6f}")

# ============================================
# (c) GRADIENT DESCENT VỚI EXACT LINE SEARCH
# ============================================
def exact_line_search(x, d, c, A, b):
    func = lambda alpha: f(x + alpha * d, c, A, b)
    res = minimize_scalar(func)
    return res.x

def gradient_descent(x0, c, A, b, p_star, max_iter=50, tol=1e-6):
    x = x0.copy()
    history = []
    for k in range(max_iter):
        g = grad_f(x, c, A, b)
        if np.linalg.norm(g) < tol:
            print(f"→ Dừng tại vòng {k+1}, gradient rất nhỏ.")
            break

        # Exact line search
        alpha = exact_line_search(x, -g, c, A, b)
        x = x - alpha * g
        fx = f(x, c, A, b)
        err = abs(fx - p_star)
        history.append((fx, err))

        print(f"Iter {k+1:2d}: f(x) = {fx:.6f}, sai số = {err:.6e}, alpha = {alpha:.4f}")
    return np.array(history)

# Chạy thuật toán
history = gradient_descent(x0, c, A, b, p_star, max_iter=50)

# ============================================
# VẼ BIỂU ĐỒ SAI SỐ
# ============================================
f_vals = history[:, 0]
errors = history[:, 1]

plt.figure(figsize=(10, 5))
plt.plot(errors, marker='o')
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('|f(x^k) - p*| (log scale)')
plt.title('Gradient Descent với Exact Line Search')
plt.grid(True)
plt.show()
```

Cho ma trận $A=\begin{bmatrix}2 & 0\\ 0 & 4 \end{bmatrix}$ và vector $b=(2;-4)$. Xét bài toán tối ưu sau

1. (2 điểm) Xác định điểm tối ưu $x^*$ và giá trị tối ưu $p^*$ của bài toán trên.

   **Gợi ý:** Tính gradient $\nabla f(x) = Ax - b$ và giải $\nabla f(x) = 0$ để tìm $x^* = A^{-1}b$.


2. (2 điểm) Sử dụng thuật toán Gradient Descent, với giá trị $x$ ban đầu là $x^{(0)}=(0;2)$, sử dụng learning rate là $0.4$ và thực hiện tối đa $30$ vòng lặp. In ra giá trị của $x^{(k)}$ và $f(x^{(k)})$ tương ứng sau mỗi vòng lặp $k$. Từ đó vẽ đồ thị biểu thị cho sai số $|f(x^{(k)})-f(p^*)|$.


3. (2 điểm) Tìm learning rate theo phương pháp exact-line search trong thuật toán Gradient Descent của bài toán trên với điểm khởi tạo là $x^{(0)}=(0;2)$. Gọi điểm cập nhật trong trường hợp này là $x^{(k)}_{\text{exact}}$, vẽ đồ thị mô tả cho sai số $|f(x^{(k)}_{\text{exact}})-f(p^*)|$.

   **Công thức exact line search cho hàm bậc hai:**
   $$\alpha_k = \frac{\|\nabla f(x^{(k)})\|^2}{\nabla f(x^{(k)})^T A \nabla f(x^{(k)})} = \frac{\|g^{(k)}\|^2}{g^{(k)T} A g^{(k)}}$$
   với $g^{(k)} = \nabla f(x^{(k)})$


```python
# 1) Xác định điểm tối ưu và giá trị tối ưu
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Định nghĩa ma trận A và vector b
A = np.array([[2, 0], [0, 4]])
b = np.array([2, -4])

# Hàm mục tiêu f(x)
def f(x):
    return 0.5 * x.T @ A @ x - b.T @ x

# Gradient của f(x)
def grad_f(x):
    return A @ x - b

# **1: Xác định điểm tối ưu x* và giá trị tối ưu p***

# Giải phương trình A*x* = b để tìm x*
x_star = np.linalg.solve(A, b)
f_star = f(x_star)

print("Phần 1:")
print(f"Điểm tối ưu x* = {x_star}")
print(f"Giá trị tối ưu p* = {f_star}\n")
```

```python
# 2) Sử dụng gradient Descent Với x0 = (0;2), learning rate là  0.4 và thực hiện tối đa  30 vòng lặp.
# Khởi tạo
x_0 = np.array([0, 2])  # Giá trị ban đầu x(0)
learning_rate = 0.4  # Learning rate cố định
max_iter = 30  # Số vòng lặp tối đa

# Danh sách lưu trữ các giá trị x(k) và f(x(k))
x_vals = [x_0]
f_vals = [f(x_0)]

# Gradient Descent
x = x_0
for k in range(max_iter):
    grad = grad_f(x)  # Tính gradient tại x(k)
    x = x - learning_rate * grad  # Cập nhật giá trị của x(k)
    x_vals.append(x)
    f_vals.append(f(x))

# In kết quả của từng vòng lặp
print("Phần 2: Gradient Descent với learning rate = 1")
for k in range(max_iter + 1):
    print(f"Vòng lặp {k}: x = {x_vals[k]}, f(x) = {f_vals[k]}")

# Vẽ đồ thị sai số |f(x^k) - f(x*)| theo số vòng lặp
errors = [abs(f_val - f_star) for f_val in f_vals]

plt.figure(figsize=(10, 5))
plt.plot(range(max_iter + 1), errors, marker='o')
plt.title("Sai số |f(x^{(k)}) - f(x^*)| qua các vòng lặp")
plt.xlabel("Số vòng lặp k")
plt.ylabel("Sai số")
plt.grid(True)
plt.show()
```

3. (2 điểm) Tìm learning rate theo phương pháp exact-line search trong thuật toán Gradient Descent của bài toán trên với điểm khởi tạo là $x^{(0)}=(0;2)$. Gọi điểm cập nhật trong trường hợp này là $x^{(k)}_{\text{exact}}$, vẽ đồ thị mô tả cho sai số $|f(x^{(k)}_{\text{exact}})-f(p^*)|$.

   **Công thức exact line search cho hàm bậc hai:**
   $$\alpha_k = \frac{\|\nabla f(x^{(k)})\|^2}{\nabla f(x^{(k)})^T A \nabla f(x^{(k)})} = \frac{\|g^{(k)}\|^2}{g^{(k)T} A g^{(k)}}$$
   với $g^{(k)} = \nabla f(x^{(k)})$

```python
# 3) Tìm learning rate theo phương pháp exact-line search
x_0 = np.array([0, 2])
max_iter = 30  # Giới hạn số vòng lặp xuống còn 20
A = np.array([[2, 0], [0, 4]])
b = np.array([2, -4])
# Hàm mục tiêu
def f(x):
    return 0.5 * x.T @ A @ x - b.T @ x

# Tính gradient
def grad_f(x):
    return A @ x - b

# Tìm điểm tối ưu p*
p_star = np.linalg.solve(A, b)
f_star = f(p_star)

# Hàm exact-line search để tìm learning rate tối ưu
def exact_line_search(grad):
    return grad.T @ grad / (grad.T @ A @ grad)  # Công thức tìm learning rate tối ưu

# Danh sách lưu trữ các giá trị x(k) và f(x(k)) với exact-line search
x_vals_exact = [x_0]
f_vals_exact = [f(x_0)]
learning_rates = []  # Danh sách lưu learning rates

x = x_0
for k in range(max_iter):
    grad = grad_f(x)  # Tính gradient tại x(k)
    lr_exact = exact_line_search(grad)  # Tìm learning rate tối ưu
    learning_rates.append(lr_exact)  # Lưu learning rate
    x = x - lr_exact * grad  # Cập nhật giá trị của x(k) với learning rate tối ưu
    x_vals_exact.append(x)
    f_vals_exact.append(f(x))

# In kết quả của từng vòng lặp với exact-line search
print("\nPhần 3: Exact-line Search")
for k in range(max_iter + 1):
    print(f"Vòng lặp {k}: x = {x_vals_exact[k]}, f(x) = {f_vals_exact[k]}")

# In ra bước nhảy tại mỗi vòng lặp
print("\nBước nhảy (learning rates) tại mỗi vòng lặp:")
for k in range(max_iter):
    print(f"Vòng lặp {k}: learning rate = {learning_rates[k]}")
```

```python
# Vẽ đồ thị sai số |f(x^k_exact) - f(x*)| theo số vòng lặp
errors_exact = [abs(f_val - f_star) for f_val in f_vals_exact]

plt.figure(figsize=(10, 5))
plt.plot(range(max_iter + 1), errors_exact, marker='o')
plt.title("Sai số |f(x^{(k)}_{exact}) - f(x^*)| qua các vòng lặp")
plt.xlabel("Số vòng lặp k")
plt.ylabel("Sai số")
plt.grid(True)
plt.show()
```
