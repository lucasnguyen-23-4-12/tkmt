### Cu 1:

 bi yu cu tnh gi tr ca pixel  $A$ trong nh input v pixel $ B$ trong nh output sau khi p dng hm bin i:

$
s = T(r) = \alpha \log_2(1 + r) + \beta
$

trong  $ \alpha$ v $ \beta $ l hng s.

### Bc 1: Xc nh cc pixel  bit v thit lp phng trnh

Ta c cc ma trn input v output:

**Ma trn Input:**
$$
\begin{bmatrix}
  & 3 &  & 3 \\
 9 &  & 15 &  \\    
  & A &  &  \\
 1 &  &  & 3 \\
\end{bmatrix}
$$

**Ma trn Output:**
$$
\begin{bmatrix}
& 5 &  & 5 \\
 &  & 11 & 9 \\
3 & 8 &  &  \\
B & 9 &  & 5 \\
\end{bmatrix}
$$

### Bc 2: Thit lp cc phng trnh da trn cc cp gi tr pixel  bit

1.  v tr (1,2): $r = 3$, $s = 5$:
   $
   5 = \alpha \log_2(1 + 3) + \beta = \alpha \cdot 2 + \beta \Rightarrow 2\alpha + \beta = 5
   $



3.  v tr (2, 3) $r = 15 $ ,$s = 11$:
   $
   11 = \alpha \log_2(1 + 15) + \beta = \alpha \cdot 4 + \beta \Rightarrow 4\alpha + \beta = 11
   $



### Bc 3: Gii h phng trnh  tm $\alpha $ v $\beta $

T phng trnh (1) v (2), ta c:
1. $ 2\alpha + \beta = 5 $
2. $ 4\alpha + \beta = 11 $



$\alpha = 3$ v $\beta = -1$

Vy ta c:
$
\alpha = 3 \quad \text{v} \quad \beta = -1
$

### Bc 4: Tm gi tr ca \( A \) v \( B \)

1. **Tm $ A $ **: Theo ma trn output, khi input l $  A $ , output s l $  8 $ .
   $ 
   8 = 3 \log_2(1 + A) - 1
   $ 
   $ 
   9 = 3 \log_2(1 + A)
   $ 
   $ 
   \log_2(1 + A) = 3
   $ 
   Chuyn t logarithm c s 2:
   $ 
   1 + A = 2^3 = 8
   $ 
   $ 
   A = 7
   $ 

2. **Tm $  B$ **: Theo ma trn input, khi input l $  1 $ , output s l $  B $ .
   $ 
   B = 3 \log_2(1 + 1) - 1 = 3 \cdot 1 - 1 = 2
  $ 

### Kt qu cui cng

- $  A = 7 $ 
- $  B = 2 $

Histogram equalization l qu trnh ci thin  tng phn ca nh bng cch phn b li mc xm ca nh.  thc hin cn bng histogram, ta thc hin cc bc sau:

#### Bc 1: Tnh histogram ca nh

- Mc xm 7: xut hin 2 ln

Ta c bng tn sut (histogram) nh sau:

| Mc xm | Tn sut (s ln xut hin) |
|---------|-----------------------------|
| 0       | 3                           |
| 1       | 4                           |
| 2       | 4                           |

#### Bc 2: Tnh xc sut xut hin ca mi mc xm
 tnh xc sut, ta chia s ln xut hin ca mi mc xm cho tng s pixel trong nh ($4 \times 4 = 16$).

| Mc xm | Tn sut | Xc sut xut hin ($p$) |
|---------|----------|--------------------------|
| 0       | 3        | $\frac{3}{16} = 0.1875$  |
| 1       | 4        | $\frac{4}{16} = 0.25$    |

#### Bc 3: Tnh phn phi tch ly (CDF) cho mi mc xm
CDF l tng ca cc xc sut t mc xm u tin n mc xm hin ti.

| Mc xm | Xc sut ($p$) | CDF                     |
|---------|----------------|-------------------------|
| 0       | 0.1875         | 0.1875                  |
| 1       | 0.25           | 0.1875 + 0.25 = 0.4375  |
| 2       | 0.25           | 0.4375 + 0.25 = 0.6875  |
| 4       | 0.0625         | 0.6875 + 0.0625 = 0.75  |
| 5       | 0.0625         | 0.75 + 0.0625 = 0.8125  |
| 7       | 0.125          | 0.8125 + 0.125 = 0.9375 |

#### Bc 4: Chun ha CDF  c mc xm mi
Ta s dng cng thc chun ha  tnh mc xm mi:

$$
\text{Mc xm mi} = \text{round}((\text{CDF} - \text{CDF ti thiu}) \times (L - 1))
$$

Vi $L = 8$ (s mc xm) v $\text{CDF ti thiu} = 0.1875$.

- Mc xm 0: $\text{round}((0.1875 - 0.1875) \times 7) = 0$
- Mc xm 1: $\text{round}((0.4375 - 0.1875) \times 7) = 2$
- Mc xm 2: $\text{round}((0.6875 - 0.1875) \times 7) = 4$
- Mc xm 4: $\text{round}((0.75 - 0.1875) \times 7) = 4$
- Mc xm 5: $\text{round}((0.8125 - 0.1875) \times 7) = 5$
- Mc xm 7: $\text{round}((0.9375 - 0.1875) \times 7) = 6$
